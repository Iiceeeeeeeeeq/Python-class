1. จงเขียนโปรแกรม วิเคราะห์พฤติกรรมการใช้จ่ายรายวัน สมมุติว่าเราได้บันทึกการใช้จ่ายของตัวเองในแต่ละวันเป็นจำนวนเงิน (บาท) เป็นเวลา 30 วัน
Process : 
    - สร้างข้อมูลใช้จ่ายสุ่มด้วย numpy
    - คำนวณค่าเฉลี่ย และส่วนเบี่ยงเบนมาตรฐาน
    - พล็อตกราฟเส้นแสดงการใช้จ่ายแต่ละวัน
    - ใส่เส้นแนวนอนแสดงค่าเฉลี่ย
    - แสดงจุดที่ใช้จ่ายสูงสุดและต่ำสุด

2. จงเขียนกราฟวิเคราะห์คะแนนสอบ 3 วิชา คือ คณิต ฟิสิกส์ เคมี

Process: 
    - สร้างข้อมูลคะแนนแบบสุ่มให้กับนักเรียน 10 คน
    - คำนวณค่าเฉลี่ยและส่วนเบี่ยงเบนของแต่ละวิชา
    - พล็อต bar chart เปรียบเทียบคะแนนเฉลี่ย 3 วิชา

3. จงเขียนกราฟวิเคราะห์แนวโน้มราคาน้ำมัน โดยสมมติว่าคุณได้รับข้อมูลราคาน้ำมันในแต่ละวันตลอด 60 วัน (เป็นระยะเวลา 2 เดือน) จากแหล่งข้อมูลภายนอก
Process: 
    - สร้างข้อมูลราคาน้ำมันแบบสุ่มที่ขึ้นๆ ลงๆ (ใช้ numpy)
    - วาดกราฟเส้นแสดงราคาน้ำมันตามวัน
    - แสดงวันราคาสูงสุดและต่ำสุดด้วยจุด

4. โรงงานแห่งหนึ่งมีต้นทุนผลิตสินค้าต่อชิ้นแบบแปรผันดังนี้:
                
                Cost(x) = 5x + 1000

    โดย  x = จำนวนสินค้าที่ผลิตต่อวัน (ตั้งแต่ 0 ถึง 500)

Process:
    - พล็อตกราฟต้นทุนต่อปริมาณการผลิต
    - แสดงช่วงที่ต้นทุนเฉลี่ยต่อชิ้นต่ำที่สุด
    - พล็อตกราฟต้นทุนเฉลี่ยต่อชิ้น

5. จงเขียนกราฟ Pie Chart แสดงข้อมูลการปล่อยคาร์บอนของกิจกรรมในชีวิตประจำวัน  (การเดินทาง, การกิน, การใช้ไฟฟ้า, ชอปปิ้ง) โดยข้อมูลเป็นดังตาราง

                activities    |   ปริมาณ CO2 (โดยประมาณ) (kg)
            ------------------|-------------------------------
                Transport     |             4.5
                Electricity   |             3.0
                Food          |             2.5
                Shopping      |             1.0
